```python
import scanpy as sc
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
ecto_hydra = sc.read("/home/gaurav/Gaurav/Berlin/Thesis/Paper/Hydra/Data/Ectoderm/adata_ecto_data.h5ad")
endo_hydra = sc.read("/home/gaurav/Gaurav/Berlin/Thesis/Paper/Hydra/Data/Endoderm/adata_endo_data.h5ad")
ic_hydra = sc.read("/home/gaurav/Gaurav/Berlin/Thesis/Paper/Hydra/Data/IC/adata_ic_data.h5ad")
```

```python
ecto_hydra.obs["cells"] = "Ectoderm"
endo_hydra.obs["cells"] = "Endoderm"
ic_hydra.obs["cells"] = "Interstitial"
```

```python
adata_hydra = ic_hydra.concatenate(ecto_hydra,endo_hydra)
```

```python
adata_hydra.obs
```

```python
sc.pp.filter_genes(adata_hydra, min_cells=20)
```

```python
sc.pp.normalize_total(adata_hydra, target_sum=1e4)
```

```python
ecto_hydra = adata_hydra[adata_hydra.obs["cells"] == "Ectoderm"]
endo_hydra = adata_hydra[adata_hydra.obs["cells"] == "Endoderm"]
ic_hydra = adata_hydra[adata_hydra.obs["cells"] == "Interstitial"]
```

```python
ic_hydra.obs["doublets"] = 0
ic_hydra.obs["female.germline"] = 0
ic_hydra.obs["male.germline"] = 0
ic_hydra.obs["spumous"] = 0
ic_hydra.obs["zymogen"] = 0
ic_hydra.obs["ic.for.tree"] = 0
ic_hydra.obs["cropped.cells"] = 0
ic_hydra.obs["outliers.nmf.full.nonendo"] = 0
ic_hydra.obs["outliers.nmf.full.nonecto"] = 0
```
