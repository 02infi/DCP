```python
epochs = pd.read_csv("/home/gaurav/Gaurav/Berlin/Paper/New_figures/Zebrafish/Trained_MesoEndoderm/Epoch.csv")
NB_loss = pd.read_csv("/home/gaurav/Gaurav/Berlin/Paper/New_figures/Zebrafish/Trained_MesoEndoderm/NB_likelihood.csv")
MMD_loss = pd.read_csv("/home/gaurav/Gaurav/Berlin/Paper/New_figures/Zebrafish/Trained_MesoEndoderm/Mmd_loss.csv")
Performance_loss = pd.read_csv("/home/gaurav/Gaurav/Berlin/Paper/New_figures/Zebrafish/Trained_MesoEndoderm/Performance_loss.csv")
```

```python
df_list = [epochs,NB_loss["Value"],MMD_loss["Value"],Performance_loss["Value"]]
df = pd.concat(df_list,axis=1)
df.columns = ["Time","Step","Epochs","NB loss","MMD loss","Total loss"]
```
